```python
from DetectModel import Detector
from DataLoader import Loader
import pandas as pd

detector = Detector()
loader = Loader()
```

```python
data_pth = 'Modeling/data/validation/Combined_LabelledData_297_전류고조파평균.json'

df = loader.make_df(data_pth)
df.index = pd.to_datetime(df.index)
X = loader.temporalize(df.values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 그래프 설정
fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)

# 데이터 업데이트 함수
def update(i):
    # 실시간으로 x (시간)과 y (상전압평균) 데이터 업데이트
    x = df.index[:i+1]
    y = df['상전압평균'][:i+1]
    
    line.set_data(x, y)
    
    ax.set_xlim(df.index.min(), df.index.max())
    ax.set_ylim(min(df['상전압평균']), max(df['상전압평균']))
    
    return line,

# 애니메이션 설정
ani = FuncAnimation(fig, update, frames=len(df), interval=1000, blit=True)

# 그래프 표시
plt.xlabel('Time')
plt.ylabel('상전압평균')
plt.title('Real-time Visualization of 상전압평균')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
